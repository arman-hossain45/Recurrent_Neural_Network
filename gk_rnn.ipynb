We want to read only clean text we remove double quatation and comma and hash we wnat to remove this in our text data so that we apply differnt types of methods to remove all of this in details

```python
# tokenize with regular expression
def tokenize(text):
  # Lowercase
  text =text.lower()

  # Remove quotes
  text = re.sub(r"['\"]","",text)
  # single quatation and double quotes remove hoye jabe

  #Remove all punctuation
  text = re.sub(r"[^a-z0-9\s]", " ", text)

  # remove extra space
  text = re.sub(r"\s+"," ",text).strip()

  #tokenize split into words
  tokens = text.split()
  return tokens
```

# Vocab forming

```python
vocab = {'<UNK>':0}
def build_vocab(row):
  print(row['question'],row['answer'])
  tokenize_question = tokenize(row['question'])
  tokenize_answer =  tokenize(row['answer'])
  #print(tokenize_question,tokenize_answer)
  merge_tokens = tokenize_question + tokenize_answer
 # print(merge_tokens)

  for token in merge_tokens:
    if token not in vocab :
      vocab[token]= len(vocab)
```

# text to indices

```python
def text_to_indices(text,vocab):
  indexed = []
  for token in tokenize(text):
    if token in vocab:
      #print(vocab[token])
      indexed.append(vocab[token])

    else :
      indexed.append(vocab['<UNK>'])
  return indexed
```

# Data set and dataloader

```python
import torch
from torch.utils.data import Dataset,DataLoader
```

```python
index= 0
text_to_indices(df.iloc[index]['question'],vocab)
```

```python
class QADataset(Dataset):

  def __init__(self,df,vocab):
    self.df = df
    self.vocab = vocab

  def __len__(self):
    return self.df.shape[0]

  def __getitem__(self,index):
   num_qus= text_to_indices(self.df.iloc[index]['question'],self.vocab)
   num_ans= text_to_indices(self.df.iloc[index]['answer'],self.vocab)
   return torch.tensor(num_qus),torch.tensor(num_ans)
```

```python
dataset = QADataset(df,vocab)
dataset[1]
```

```python
dataloader = DataLoader(dataset,batch_size=1,shuffle=True)
```

```python
for question,answer in dataloader:
  print(question,answer)
```

# RNN architeccture

```python
import torch.nn as nn

class RNN(nn.Module):
  def __init__(self,vocab_size,embedding_dim=50,hidden_size=64):
    super().__init__()
    self.embedding = nn.Embedding(vocab_size,embedding_dim)
    self.rnn = nn.RNN(embedding_dim,hidden_size,batch_first=True)
    self.fc = nn.Linear(hidden_size,vocab_size)

  def forward(self,question):
    embedded = self.embedding(question)
    _, final_hidden_state = self.rnn(embedded)
    # final_hidden_state has shape (num_layers * num_directions, batch_size, hidden_size)
    # For a single-layer, unidirectional RNN, this is (1, batch_size, hidden_size)
    # We need to squeeze out the first dimension before passing to the linear layer
    return self.fc(final_hidden_state.squeeze(0)) # (batch_size, vocab_size)
```

```python
model=RNN(vocab_size=len(vocab))
```

#Tranning loop

```python


criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.0001)
epochs = 50


for epoch in range(epochs):
    total_loss = 0
    for question, answer in dataloader:
        optimizer.zero_grad()
        output = model(question)
        # Fix: take only the first answer token as target → shape (1,)
        target = answer[0][0].unsqueeze(0)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1:3d}/{epochs}  |  Loss: {total_loss:.4f}")
```

```python


def predict(question_text,model,vocab,idx_to_word):
  model.eval()

  with torch.no_grad():
    indices = text_to_indices(question_text,vocab)

    if not indices:
      return ""

    x = torch.tensor(indices).unsqueeze(0)  #batch size, sequence legth

    pred = torch.argmax(model(x),dim=1).item()


  return idx_to_word.get(pred,"")
```
